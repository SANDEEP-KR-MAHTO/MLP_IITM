## Finding Outliers

```python
lower_limit = data['ram_capacity'].quantile(0.01)
lower_limit
```

```python
upper_limit = data['ram_capacity'].quantile(0.99)
upper_limit
```

```python
data[(data['ram_capacity'] < lower_limit) | (data['ram_capacity'] > upper_limit)]
```

## Trimming

```python
trimmed_data = data[(data['ram_capacity'] <= upper_limit) & (data['ram_capacity'] >= lower_limit)]
trimmed_data
```

```python
# Comparing

plt.figure(figsize=(10,8))
plt.subplot(2,2,1)
sns.histplot(data['ram_capacity'], kde= 'True', bins=60)

plt.subplot(2,2,2)
sns.boxplot(data['ram_capacity'], orient='h')

plt.subplot(2,2,3)
sns.histplot(trimmed_data['ram_capacity'], kde='True', bins=60)

plt.subplot(2,2,4)
sns.boxplot(trimmed_data['ram_capacity'], orient='h')

plt.show()
```

## Capping

```python
capped_data = data.copy()

capped_data['ram_capacity'] = np.where(
    capped_data['ram_capacity'] > upper_limit,
    upper_limit,
    np.where(
        capped_data['ram_capacity'] < lower_limit,
        lower_limit,
        capped_data['ram_capacity']
    )
)
```

```python
# Comparing

plt.figure(figsize=(10,5))
plt.subplot(2,2,1)
sns.histplot(data['ram_capacity'], kde= 'True', bins=50)

plt.subplot(2,2,2)
sns.boxplot(data['ram_capacity'], orient='h')

plt.subplot(2,2,3)
sns.histplot(capped_data['ram_capacity'], kde= 'True', bins=50)

plt.subplot(2,2,4)
sns.boxplot(capped_data['ram_capacity'], orient='h')

plt.show()
```
