## 1. Feature Scaling

#### Feature Scaling is a process of transforming the feature values so that feature values have similar scale. It helps ML Algorithms to converge quickly. Some Algorithms like SVM, K-NN, K-means compute euclidean distance among the data points get impacted.

```python
# Lets create a sample of data points having different range of feature values.
```

```python
x1 = np.random.randint(1,20, size=(20,1))
x2 = np.random.randint(10,200, size=(20,1))
x3 = np.random.randint(100, 2000, size=(20,1))
X = np.concatenate((x1,x2,x3), axis=1)
X
```

```python
# Create a new plot

plt.figure(figsize=(10,8))

#plot the features
plt.plot(x1)
plt.plot(x2)
plt.plot(x3)

plt.legend(['x1', 'x2', 'x3'])

plt.show()
```

```python
# mean and standard deviation
```

```python
X.mean(axis=0)
```

```python
np.mean(X, axis=0)
```

```python
np.std(X, axis =0)
```

### **1. Normalization**

#### Normalization is a feature scaling technique in which feature values are transformed in the range of [0,1]. This technique is called MinMax Scaling.

```python
from sklearn.preprocessing import MinMaxScaler
MS = MinMaxScaler()
```

```python
ms = MS.fit(X)
```

```python
X_normalized = ms.transform(X)
X_normalized
```

```python
# Mean and standard deviation of features
```

```python
X_normalized.mean(axis=0)
```

```python
X_normalized.std(axis=0)
```

### **2. Standardization**

#### Standardization is another technique to transform the feature values into mean of data points close to zero and unit standard deviation.

```python
from sklearn.preprocessing import StandardScaler
SS = StandardScaler()
```

```python
ss = SS.fit(X)
```

```python
X_standardize = ss.transform(X)
X_standardize
```

```python
X_standardize.mean(axis=0)
```

```python
X_standardize.std(axis=0)
```

## 2.Data Imputation

#### Data Imputation is the procedure to fill up the missing values in the dataset. We can do this by replacing missing values with
* Feature's mean value
* Feature's median value
* Feature's mode value
* Some constant value

```python
from sklearn.impute import SimpleImputer
```

## 3. Handling Categorical Features

Handling Categorical Features means converting categories present in the dimension into numericals as machine or algorithms don't understand categories or words, it only understands numbers or vectors.

One of the most common approach is One-Hot Encoding that we will see in this colab

Other than that Ordinal and Label Encoding is also there,Label Encoding is used for target column and Ordinal is used for Columns that contains categories which have order in it.

```python
from sklearn.preprocessing import OneHotEncoder
```

## 4. Handling Imbalance Data

Imbalanced datasets are those where one class is very less represented than the other class. This kind of data results in less efficient ML algorithms.

image.png

4.1 Create imbalance synthetic dataset

```python
from sklearn.datasets import make_classification
```

```python
X, y = make_classification(n_samples=10000, n_features=2, n_redundant=0, n_clusters_per_class=1, weights=[0.99], flip_y=0, random_state=100)
```

The details of the above parameters are as follows:

* n_samples (default = 100): number of required samples

* n_features (default = 20): number of features in the dataset

* n_redundant (default = 2): number of redundant features, i.e., random linear combinations of the informative features.

* n_clusters_per_class (default = 2): number of clusters per class

* weights (default = None): array of size n_classes or n_classes-1 storing the proportion of samples assigned to each class. (None means the classes will be balanced). If size of array is n_classes-1, then the size of last class is inferred automatically. Note: if flip_y is not zero, then the class ratios will not be exact.

* flip_y (default=0.01): fraction of samples whose class is assigned randomly. More this value, more difficult to get a clean classification.

* random_state (default=None): random number generation for dataset creation. Pass an int for reproducible output.

* n_classes (default=2): number of classes (or labels) of the classification problem.

```python
#It allows you to count the items in the iterable list
from collections import Counter
count = Counter(y)
count
```

```python
colors = np.where(y==1, 'k', 'g')

plt.figure(figsize=(10,8))
plt.scatter(X[:,0], X[:,1], c = colors, alpha=0.2)
plt.show()
```

4.2 Undersampling

Undersampling refers to the sampling from the majority class in order to keep only a parts of these data points.

It may be carried out using **RamdonUnderSampler** class from **imblearn** library.

```python
from imblearn.under_sampling import RandomUnderSampler
```

```python
Under_sampler = RandomUnderSampler(sampling_strategy = 'majority')
X2, y2 = Under_sampler.fit_resample(X,y)
counter = Counter(y2)
counter
```

```python
color = np.where(y2==1, 'k', 'g')

plt.figure(figsize=(10,8))
plt.scatter(X2[:,0], X2[:,1], c = color, alpha=0.2)
plt.show()
```

4.3 Oversampling

Oversampling refers to replicating some points from the minority class in order to increase the cardinality of the minority class. This might consist of either replicating or generating synthetic data from the minority class.

SMOTE(Synthetic Minority Oversampling Technique) is a popular technique for oversampling.

It is available under imblearn library.

```python
from imblearn.over_sampling import SMOTE
```

```python
oversampler = SMOTE()
```

```python
X1, y1 = oversampler.fit_resample(X,y)
```

```python
Counter(y1)
```

```python
colors = np.where(y1==1, 'k', 'g')

plt.figure(figsize=(10,8))
plt.scatter(X1[:,0], X1[:,1], c = colors, alpha=0.2)
plt.show()
```
