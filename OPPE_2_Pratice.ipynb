(Common Instructions for Q12, Q13 and Q14)

Suppose we train a decision tree classifier using the following two approaches:

    Approach 1. Do not grow the full size tree by setting the parameters as : max depth of the tree to be 10. minimum samples in a node to split to be 50.

    Approach 2. Use cost complexity pruning by setting the cost complexity parameter as 0.01.

NOTE:


Use random state to be 64 in both approaches. Use entropy criterion in both approaches.




## Q12. Which approach gives the better accuarcy on the test dataset?

```python
#Approach 1.

from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
app1= DecisionTreeClassifier(random_state=64, criterion='entropy', max_depth= 10, min_samples_split= 50)

app1.fit(X_train, y_train)
y_pred_app1= app1.predict(X_test)
score_1= accuracy_score(y_test, y_pred_app1)
print(score_1)



#Approach 2.

app2= DecisionTreeClassifier(random_state=64, criterion= 'entropy', ccp_alpha=0.01)
app2.fit(X_train, y_train)
y_pred_app2= app2.predict(X_test)
score2=accuracy_score(y_test, y_pred_app2)
print(score2)
```

## Q13. What is the difference (absolute difference) in True Positives (TP) between these two approaches on test datasets?

```python
from sklearn.metrics import confusion_matrix
cm1= np.array(confusion_matrix(y_test, y_pred_app1))
cm2=confusion_matrix(y_test, y_pred_app2)
```

```python
diff = abs(cm1[1,1]- cm2[1,1])
print(diff)
```

## Q14. Find the sum of number of nodes of both trees (trees obtained using approach 1 and appoach 2).
