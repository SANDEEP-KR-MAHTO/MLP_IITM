```python
#Finding the IQR

percentile25 = df['price'].quantile(0.25)
percentile75 = df['price'].quantile(0.75)
```

```python
IQR = percentile75 - percentile25
```

```python
percentile75
```

```python
IQR
```

```python
upper_limit = IQR + percentile75 * 1.5
lower_limit = IQR - percentile25 * 1.5
```

## Finding Outliers

```python
df[df['price'] < lower_limit]
```

```python
df[df['price'] > upper_limit]
```

## Trimming

```python
df[(df['price'] > upper_limit) | (df['price'] < lower_limit)]
```

```python
trimmed_df = df[(df['price'] < upper_limit) & (df['price']> lower_limit)]
trimmed_df
```

## Comparing

```python
plt.figure(figsize=(16,8))
plt.subplot(2,2,1)
sns.histplot(df['price'], kde= 'True', bins=60)

plt.subplot(2,2,2)
sns.boxplot(df['price'], orient='h')

plt.subplot(2,2,3)
sns.histplot(trimmed_df['price'], kde='True', bins=60)


plt.subplot(2,2,4)
sns.boxplot(trimmed_df['price'], orient = 'h')
```

## Capping

```python
capped_df = df.copy()
capped_df['price'] = np.where(
    capped_df['price'] > upper_limit,
    upper_limit,
    np.where(
        capped_df['price'] < lower_limit,
        lower_limit,
        capped_df['price']
    )
)
```
