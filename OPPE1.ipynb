### Rules for Imputing the missing(NaN) or unknown ("?") values:

1.compute the instructed statistical values for different columns in train dataset only to repalce missing(NaN) and unknown("?") values respectively.

2.Replace the missing(NaN) and unknown("?") values from the train and test dataset with the instructed statistical values computed using the train dataset only.

3.Ignore the missing and unknown values while calculating the statistical values.

4.Replace missing values(NaN) with the MOST FREQUENT value of the Locality feature

5.Repalce missing values(NaN) with the MEDIAN value of the Estimated Value feature

6.Replace missing values(NaN) with the MEAN value of the carpet_area feature

7.Replace Unknown values("?") with the MOST FREQUENT value of the Property.

Write the answers related to the above imputation in below questions respectively.

```python
X_trainR=X_train.replace('?', np.nan)
X_testR=X_test.replace('?', np.nan)
```

```python
X_trainR.head()
```

```python
X_trainR['carpet_area'].mean()
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
```

```python
ct= ColumnTransformer([
      ('locality', SimpleImputer(strategy='most_frequent'), ['Locality']),
      ('estimated_value', SimpleImputer(strategy='median'), ['Estimated Value']),
      ('carpet_area', SimpleImputer(strategy='mean'), ['carpet_area']),
      ('propety', SimpleImputer(strategy='most_frequent'), ['Property'])],
                      remainder='passthrough', verbose_feature_names_out=False)

XtrainCT=ct.fit_transform(X_trainR)
XtestCT=ct.transform(X_testR)
```

```python
XtrainCT    #Converted into array
```

```python
XtestCT
```

```python
#Again have to convert into DataFrame
Xtrainp=pd.DataFrame(XtrainCT, columns=ct.get_feature_names_out())
Xtestp = pd.DataFrame(XtestCT, columns=ct.get_feature_names_out())
```

(Common Instructions for Question 14 and 15)


* create a pipeline of the PolynomialFeatures(interaction_only=True) as transformer and Lasso as an estimator.

* Use GridSearchCV for tuning the hyperparameters of the created pipeline on training dataset.

1. Keep polynomial degree as : [1,2]

2. lasso alpha value to be taken as : [10,100,1000,10000]

3. scoring : neg_mean_absolute_error.

4. cv = 5

5. n_jobs = -1 (negative one) [it helps in using all the computational power to run this job]

(Note: Kindly ignore the warning.)


```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import PolynomialFeatures
from sklearn.model_selection import GridSearchCV

pipe=Pipeline([
     ('ploy', PolynomialFeatures(interaction_only=True)),
     ('lasso', Lasso())
 ])


param_grid=[
     {'ploy__degree': [1,2], 'lasso__alpha': [10,100,1000,10000]}
 ]


search_grid=GridSearchCV(pipe, param_grid=param_grid, scoring='neg_mean_absolute_error', cv=5, n_jobs=-1)
search_grid.fit(X_trainpp, y_train)
```
