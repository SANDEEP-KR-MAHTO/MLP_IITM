Common Instructions for Question 8 to 9


  * Rules for Imputing the missing(NaN) or Unknown values:


  * Calculating statistical values (such as mean, median, mode) for each column in the training dataset.

  * Applying these calculated statistical values to replace missing (NaN) and unknown values in both the training and test datasets.

  * Ensuring that the calculation of statistical values excludes any rows containing missing or unknown values.

  * After imputation store the dataframe in another variable imputed_df

  * Replace missing values in the Agefeature with the Median.

  * Replace missing values in the Annual Income feature with the Mean.

  * Replace missing values in the Number of Dependents feature with the Most Frequent.

  * Replace missing values in the Health Score feature with the Median.

  * Replace missing values in the Credit Score feature with the Constant value 0 (zero as integer data type).

  * Replace missing values in the Customer Feedback feature with the Most Frequent value.

  * Write the answers related to the above imputation in below questions respectively.

```python
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer

```

```python
col=ColumnTransformer(
    transformers=[
        ('age', SimpleImputer(strategy='median'), ['Age']),
        ('annual_income', SimpleImputer(strategy='mean'), ['Annual Income']),
        ('no_of_dependent', SimpleImputer(strategy='most_frequent'), ['Number of Dependents']),
        ('health_score', SimpleImputer(strategy='median'), ['Health Score']),
        ('credit_score', SimpleImputer(strategy='constant', fill_value=0), ['Credit Score']),
        ('customer_feedback', SimpleImputer(strategy='most_frequent'), ['Customer Feedback'])
           ], remainder='passthrough', verbose_feature_names_out=False)


imputed_train_data=col.fit_transform(X_train)
imputed_test_data=col.transform(X_test)
```

```python
X_trainp=pd.DataFrame(imputed_train_data, columns=col.get_feature_names_out())
X_testp=pd.DataFrame(imputed_test_data, columns=col.get_feature_names_out())
```


Apply preprocessing on features of train and test datasets.

   * After applying the above imputation steps there should not be any missing or unknown values present in the train and test dataset.

   * Fitting (learning) should be done only on train dataset after above imputation is done.

   * Learn transformers' parameters using training set only and then apply them on train & test sets.

For Categorical Features

   * Ordinally Encode "Education Level", "Exercise Frequency" features because there are too many categories and to handle dimensionality of features.


Nominal Features
   * One-Hot Encode "Gender", "Smoking Status", "Location", "Policy Type", "Customer Feedback", "Property Type" features and keep drop_first = True.


Scaling Features
  * Scale all the features (transformed categorical and numerical) of the feature matrix using the StandardScaler

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import OrdinalEncoder
from sklearn.preprocessing import StandardScaler
```

```python
ct= ColumnTransformer(
    transformers=[
        ('ordinal_encoder', OrdinalEncoder(), ['Education Level', 'Exercise Frequency']),
        ('nomial_encoder', OneHotEncoder(sparse_output=False, drop='first'), ['Gender', 'Smoking Status', 'Location', 'Policy Type', 'Customer Feedback', 'Property Type'])
    ], remainder='passthrough', verbose_feature_names_out=False
)

colTran=ct.set_output(transform='pandas')
colTran
```

```python
X_trainpp=colTran.fit_transform(X_trainp)
X_testpp=colTran.transform(X_testp)
```

```python
X_trainpp.head()
```

```python
ss=StandardScaler().set_output(transform='pandas')
X_trainppp=ss.fit_transform(X_trainpp)
X_testppp=ss.transform(X_testpp)
```
