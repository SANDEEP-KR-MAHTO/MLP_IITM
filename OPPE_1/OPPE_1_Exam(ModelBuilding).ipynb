(Common Instructions for Q.7 and Q.8)

Create a pipeline of the PCA() as transformer and Lasso as an estimator.

Use GridSearchCV for tuning the hyperparameters of the created pipeline on training dataset.






  * Values of n_components for PCA to be [0.9, 0.95]

  * lasso alpha value to be taken as : [10, 1, 0.01, 0.001]

  * scoring : neg_mean_absolute_error.

  * Use 5 fold cross validation.

  * n_jobs = -1 (negative one) [it helps in using all the computational power to run this job]

(Note: Kindly ignore the warning.)

```python
from sklearn.pipeline import Pipeline
from sklearn.decomposition import PCA
from sklearn.linear_model import Lasso
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import root_mean_squared_error
```

```python
pipe=Pipeline([
    ('pca', PCA()),
    ('lasso', Lasso())
])

param_grid=[
    {'pca__n_components':[0.9, 0.95], 'lasso__alpha': [10,1,0.01,0.001],}
]

```

```python
search_grid= GridSearchCV(pipe, param_grid, scoring='neg_mean_absolute_error', cv=5, n_jobs=-1)
```

```python
search_grid.fit(X_train, y_train)
```

```python
search_grid.best_params_
```

## Q7. If we fit the pipeline on the training dataset, what will be the Root Mean Squared Error on the test dataset?

```python
y_pred_pipe= search_grid.predict(X_test)
root_mean_squared_error(y_test, y_pred_pipe)
```


Create a pipeline of the PolynomialFeatures as transformer and Lasso as an estimator with the following parameters:

For PolynomialFeatures:

	interaction_only = False
	 degree = 2

For Lasso:

	 alpha = 1
	 warm_start = True
	 random state as 0

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import Lasso
from sklearn.metrics import mean_squared_error
from sklearn.pipeline import Pipeline
```

```python
pipe=Pipeline([
    ('poly', PolynomialFeatures(interaction_only=False, degree=2)),
    ('lasso', Lasso(alpha=1, warm_start=True, random_state=0))
])
```

```python
pipe.fit(X_train, y_train)
```

## Q9. Fit the pipeline on the training dataset and find the Mean Square Error on the test dataset.

```python
y_pred_pipe= pipe.predict(X_test)
mean_squared_error(y_test, y_pred_pipe)
```

Create a pipeline for feature selection using f_regression as transformer and SGDRegressor as an estimator with the following parameters:

For feature selection:

	select best 5 features based on f_regression score

For SGDRegressor:

	random state as 42

```python
from sklearn.pipeline import Pipeline
from sklearn.linear_model import SGDRegressor
from sklearn.feature_selection import SelectKBest, f_regression
```

```python
pipe=Pipeline([
    ('feature_selection', SelectKBest(f_regression, k=5)),
    ('sgd', SGDRegressor(random_state=42))
])

```

## Q10. Fit the pipeline on the training dataset and select the option that correctly displays the feature names in descending order based on their f_regression scores.

```python
pipe[0].feature_names_in_
```

```python
pipe[0].scores_
```
