## Trimming

```python
trimed_data = data[(data['V22'] > -2.025754709810713) & (data['V22'] < 1.812156167317039)]
trimed_data
```

```python
# Approach 2

#Calculating Z-score

data['V22_zscore'] = (data['V22'] - data['V22'].mean())/ (data['V22'].std())
```

```python
data[(data['V22_zscore'] < 3) & (data['V22_zscore'] >-3)]
```

## Capping

```python
upper_limit = data['V22'].mean() + 3 * data['V22'].std()
lower_limit = data['V22'].mean() - 3 * data['V22'].std()
```

```python
data['V22'] = np.where(
                data['V22'] > upper_limit,
                upper_limit,
                np.where(
                    data['V22'] < lower_limit,
                    lower_limit,
                    data['V22']
                        )
                )
```
